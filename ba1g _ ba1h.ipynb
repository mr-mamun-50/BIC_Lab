```python
def hamming_distance(p, q):
    cnt = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            cnt += 1

    return cnt

def appr_pattern_match(pattern, string, d):
    matches = []

    p_len = len(pattern)

    for i in range(len(string) - p_len + 1):
        substring = string[i:i+p_len]
        distance = hamming_distance(pattern, substring)

        if distance <= d:
            matches.append(i)

    return matches


pattern = input()
string = input()
d = input()

result = appr_pattern_match(pattern, string, int(d))
print(" ".join(map(str, result)))
```
