```python
text = input()
k = 4
kmers = {} # {ACGT:2, GGTT:2, ACTG:3, GTCA:1}

for i in range(len(text) - (k+1)):
	pattern = text[i:i+k]
	if pattern in kmers.keys():
		kmers[pattern] += 1
	else:
		kmers[pattern] = 1

printMax = "" # ACTG
max = 0
for index,val in kmers.items():
	if val > max:
		printMax = index
		max = val
	else:
		if val == max:
			printMax += " " + index

print(printMax)
# print(kmers)
```
